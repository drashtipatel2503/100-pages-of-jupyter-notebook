```python
def getsum(y,df,z,ps):
    s=0
    for i in range(len(df)):
        if df[y][i]==z:
            s+=df[ps][i]
    return(s)
            
```

```python
Office_suplies_netsales=getsum('Category',df,'Office Supplies', 'Sales')
```

```python
Furniture_netsales=getsum('Category',df,'Furniture', 'Sales')
```

```python
Technology_netsales=getsum('Category',df,'Technology', 'Sales')
```
